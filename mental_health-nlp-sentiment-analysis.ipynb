
# 3. Text Preprocessing 

```python
# Function to clean text
import string
def clean_text(text):
    """
    Cleans text by:
    - Lowercasing
    - Removing URLs
    - Removing Twitter handles
    - Removing punctuation and special characters
    - Stripping extra whitespace
    """
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'http\S+|www\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove Twitter handles
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra whitespace
    return text

# Function to tokenize text
def tokenize_text(text):
    """
    Tokenizes text using word_tokenize from NLTK.
    """
    return word_tokenize(text)

# Function to stem tokens
def stem_tokens(tokens):
    """
    Stems a list of tokens using PorterStemmer.
    Returns a single string of stemmed tokens.
    """
    stemmer = PorterStemmer()
    return ' '.join(stemmer.stem(str(token)) for token in tokens)

# Main function to preprocess DataFrame
def preprocess_statements(df, column_name='statement'):
    """
    Applies text cleaning, tokenization, and stemming to a DataFrame.
    
    Parameters:
        df (pd.DataFrame): Input DataFrame.
        column_name (str): Column name of the statements to preprocess.

    Returns:
        pd.DataFrame: Updated DataFrame with cleaned, tokenized, and stemmed text.
    """
    # Clean text
    df[column_name] = df[column_name].apply(clean_text)

    # Tokenize text
    df['tokens'] = df[column_name].apply(tokenize_text)

    # Stem tokens
    df['tokens_stemmed'] = df['tokens'].apply(stem_tokens)

    return df

# usage:
df = preprocess_statements(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from imblearn.over_sampling import RandomOverSampler
from scipy.sparse import hstack

def preprocess_and_resample(df):
    # 1. Prepare features and target variable
    X = df[['tokens_stemmed', 'num_of_characters', 'num_of_sentences']]
    y = df['status']
    
    # 2. Label encode target variable
    lbl_enc = LabelEncoder()
    y = lbl_enc.fit_transform(y.values)
    
    # 3. Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
    
    # 4. TF-IDF Vectorization on 'tokens_stemmed'
    vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=50000)
    X_train_tfidf = vectorizer.fit_transform(X_train['tokens_stemmed'])
    X_test_tfidf = vectorizer.transform(X_test['tokens_stemmed'])
    
    # 5. Extract numerical features (characters and sentences)
    X_train_num = X_train[['num_of_characters', 'num_of_sentences']].values
    X_test_num = X_test[['num_of_characters', 'num_of_sentences']].values
    
    # 6. Combine TF-IDF features with numerical features
    X_train_combined = hstack([X_train_tfidf, X_train_num])
    X_test_combined = hstack([X_test_tfidf, X_test_num])
    
    # 7. Print number of feature words in the TF-IDF vectorizer
    print('Number of feature words: ', len(vectorizer.get_feature_names_out()))
    
    # 8. Apply Random Over-Sampling on the vectorized data to balance class distribution
    ros = RandomOverSampler(random_state=101)
    X_train_resampled, y_train_resampled = ros.fit_resample(X_train_combined, y_train)
    
    return X_train_resampled, X_test_combined, y_train_resampled, y_test

#  usage:
X_train_resampled, X_test_combined, y_train_resampled, y_test = preprocess_and_resample(df)
```

# 5. Model Traning and Evaluation 

```python
# classifiers dictionary
classifiers = {
    'Bernoulli Naive Bayes': BernoulliNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=101),
    'Logistic Regression': LogisticRegression(solver='liblinear', random_state=101)
}

#  hyperparameter grids
param_grids = {

    'Bernoulli Naive Bayes': {
        'alpha': [0.1, 0.5, 1.0],
        'binarize': [0.0, 0.5, 1.0]
    },
    'Decision Tree': {
        'max_depth': [3, 5, 9],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 5],
        'max_features': ['auto', 'sqrt', None]
    },
    'Logistic Regression': {
        'C': [0.1, 1, 10],
        'penalty': ['l1', 'l2'],
        'max_iter': [100, 200, 500]
    }
}

# Initialize a list to store accuracy scores for each classifier
accuracy_scores = []

# Iterate over each classifier and its name in the classifiers dictionary
for name, clf in classifiers.items():
    print(f"Performing GridSearchCV for {name}...")
    
    # Initialize GridSearchCV
    grid_search = GridSearchCV(estimator=clf, param_grid=param_grids[name], cv=3, scoring='accuracy', verbose=1)
    
    # Fit the GridSearchCV model
    grid_search.fit(X_train_resampled, y_train_resampled)
    
    # Get the best model from the grid search
    best_model = grid_search.best_estimator_
    
    # Make predictions with the best model
    y_pred = best_model.predict(X_test_combined)
    
    # Calculate the accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"Best hyperparameters for {name}: {grid_search.best_params_}")
    print(f"Best accuracy score for {name}: {grid_search.best_score_}")
    
    # Compute the confusion matrix for the predictions
    labels = lbl_enc.classes_
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(classification_report(y_test, y_pred, target_names=labels))
    
    # Plot the confusion matrix using a heatmap
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')  # Label for x-axis
    plt.ylabel('Actual')     # Label for y-axis
    plt.title(f'Confusion Matrix for {name}')  # Title for the heatmap
    plt.show()  # Display the heatmap
    
    # Append the accuracy score to the list
    accuracy_scores.append(accuracy)

# print the list of accuracy scores
print("\nAccuracy scores for each classifier after GridSearchCV:")
for name, score in zip(classifiers.keys(), accuracy_scores):
    print(f"{name}: {score}")
```

```python
# Create a DataFrame to store classifier names and their corresponding accuracy scores
accuracies_df = pd.DataFrame({'Classifier': classifiers.keys(), 'Accuracy': accuracy_scores}).sort_values('Accuracy', ascending=False)

plt.figure(figsize=(12, 8))
palette = dict(zip(accuracies_df['Classifier'], colors[:4]))

# Create a bar plot to visualize the accuracy of each classifier
sns.barplot(x='Classifier', y='Accuracy', data=accuracies_df, palette=palette)

plt.title("Classifier Accuracy Comparison")
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
# source:https://www.kaggle.com/code/annastasy/mental-health-sentiment-analysis-nlp-ml
```
